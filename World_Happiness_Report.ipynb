```python
def url_to_df(url):
  '''
  takes a url to a csv as input and returns in a ready to read format
  '''
  url ='https://drive.google.com/uc?id=' + url.split('/')[-2]
  
  return url
```

```python
import pycountry
import pandas as pd

def findCountryAlpha3 (country_name):
  '''
  Takes country name as input and returns the 3 letter alpha numeric
  country code 
  '''
  try:
    return pycountry.countries.get(name=country_name).alpha_3
  except:
    return ("not founded!")

# Adding 3 letter alpha numeric country code to the dataframe for choropleth map
df['country_alpha_3'] = df.apply(lambda row: findCountryAlpha3(row.Country) , axis = 1)
```

```python
df.head()
```

```python
import plotly.graph_objects as go
import pandas as pd

df_2011 = temp_gdp_df = df[df.year == '2021']

fig = go.Figure(data=go.Choropleth(
    locations = df['country_alpha_3'],
    z = df['Happiness Score'],
    text = df['Country'],
    colorscale = 'Plasma',
    autocolorscale=False,
    reversescale=False,
    marker_line_color='darkgray',
    marker_line_width=0.5,
    colorbar_title = 'Happiness Score',
))

fig.update_layout(
    title_text='2021 Happiness Score',
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    )
)

fig.show()
```
